```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://data-class-mars-challenge.s3.amazonaws.com/Mars/index.html'
browser.visit(url)
```

```python
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
table = html_soup.find('table', class_='table')
for t in table:
    print(t.text) 
```

```python
mars_temp_data = []

data_rows = table.find_all('tr', class_='data-row')

# table_head = table.find('tr').find_all('th')
# heading = [x.text.strip() for x in table_head]

for row in data_rows:
    values = [x.text.strip() for x in row.find_all('td')]
    data = {
        "id": int(values[0]),
        "terrestrial_date": values[1],
        "sol": int(values[2]),
        "ls": int(values[3]),
        "month": int(values[4]),
        "min_temp": float(values[5]),
        "pressure": float(values[6]),
    }
    
    # data = {heading[i]: values[i] for i in range(len(heading))}
    mars_temp_data.append(data)

mars_temp_data
```

```python
mars_temp_data_df=pd.DataFrame(mars_temp_data)
mars_temp_data_df
```

```python
mars_temp_data_df['terrestrial_date'] = pd.to_datetime(mars_temp_data_df['terrestrial_date'])
```
